```python
x=2
lr=0.012
precision=0.000001
previous_step_size=1
max_iter=10000
iter=0
gm=lambda x:(x+3)**2
```

```python
gd=[]
```

```python
while precision<previous_step_size and iter<max_iter:
    prv=x
    x=x-lr*gm(prv)
    previous_step_size=abs(x-prv)
    iter=iter+1
    print(iter,x)
    gd.append(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(gd)
```
